5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Figure3D:
  def get_volume(self):
    return 0

  def __add__(self, other):
    if isinstance(other, Figure3D):
      return self.get_volume() + other.get_volume()
    else:
      raise TypeError
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
class Cylinder(Figure3D):
  def __init__(self, height, radius):
    self.height = height
    self.radius = radius

  def get_volume(self):
    volume = 3.14 * self.radius * self.radius * self.height
    return volume

class Cube(Figure3D):
  def __init__(self, side):
    self.side = side

  def get_volume(self):
    volume = self.side ** 3
    return volume

cylinder = Cylinder(3, 4)
cube = Cube(4)

print(cylinder + cube)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self, id,  balance = 0):
    self.__balance = balance
    self.__id = id

  @property
  def balance(self):
    return self.__balance

  def deposit(self, amount):
        if amount <= 0:
            raise ValueError("you can't do this")
        self.__balance += amount
        return self.__balance

  def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("you can't do this")
        if amount > self.__balance:
            raise ValueError("Not enough money")
        self.__balance -= amount
        return self.__balance
```

```python
wallet = EWallet(1, 555)


print("Your balance", wallet.deposit(55))
print("Your balance", wallet.withdraw(70))
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name, price):
    self.name = name
    self.price = price

class ShoppingCart:
  def __init__(self):
    self.items = []

  def add_item(self, item):
    self.items.append(item)

  def delete_item(self, item):
    self.items.remove(item)

  def show_shopping_cart(self):
    if not self.items:
      return None
    for item in self.items:
      return item.name
```

```python

bread = Item("bread", 15)
milk = Item("milk", 30)
water = Item("water", 5)

shopping_cart = ShoppingCart()

shopping_cart.add_item(bread)
shopping_cart.add_item(milk)
shopping_cart.add_item(water)

shopping_cart.delete_item(water)

shopping_cart.show_shopping_cart()
```

10. (13б). Створіть клас Vehicle з атрибутами brand (manufacturer) та production_age. Створіть ще один клас Engine з такими атрибутами, як type (ex. petrol) та size (volume, litres).
```python
class Vehicle:
  def __init__(self, brand, production_age, **kwargs):
    super().__init__(**kwargs)
    self.brand = brand
    self.production_age = production_age

class Engine:
  def __init__(self, tyype, size, **kwargs):
    super().__init__(**kwargs)
    # tyype з двома у, щоб не використовувати вбудовану функцію
    self.tyype = tyype
    self.size = size

class Car(Vehicle, Engine):
  def __init__(self, brand, production_age, tyype, size):
    super().__init__(
        brand = brand,
        production_age = production_age,
        tyype = tyype,
        size = size
    )
```

```python
car = Car("BMW", 2020, "petrol", 2.0)

print(Car.__mro__)
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:
  def __init__(self):
    self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    self.current = 0
    self.limit = len(self.days)

  def __iter__(self):
    return self

  def __next__(self):
    if self.current < self.limit:
      val = self.days[self.current]
      self.current += 1
      return val
    else:
      raise StopIteration

week = WeekDays()
for day in week:
    print(day)
```
